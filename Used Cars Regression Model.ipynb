```python
import tensorflow_decision_forests as tfdf
from tensorflow_decision_forests.keras import Task
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('used_car_cleaned.csv')
indexPrice1 = df[ df['car_price'] <= 15000].index
df.drop(indexPrice1,inplace=True)
indexPrice2 = df[df['car_price'] >= 200000].index
df.drop(indexPrice2,inplace=True)

indexDrive2 = df[df['car_driven'] <= 500].index
df.drop(indexDrive2,inplace=True)
```

```python
train , test= train_test_split(df, test_size=0.3,random_state=93)
train_data = tfdf.keras.pd_dataframe_to_tf_dataset(train,label = "car_price",task = Task.REGRESSION)
test_data = tfdf.keras.pd_dataframe_to_tf_dataset(test,label = "car_price",task =Task.REGRESSION)
```

```python
model_forest = tfdf.keras.RandomForestModel(task = Task.REGRESSION, max_depth = 30,min_examples = 1,num_trees = 500)
model_gradient = tfdf.keras.GradientBoostedTreesModel(task = Task.REGRESSION, max_depth = 30,min_examples = 7)
model_forest.compile(metrics=['mae', 'mse'])
model_gradient.compile(metrics=['mae', 'mse'])
```

```python
model_forest.fit(train_data)
model_gradient.fit(train_data)
```

```python
model_forest.summary()
```

```python
model_gradient.summary()
```

```python
import math
results_forest = model_forest.evaluate(test_data)
results_gradient = model_gradient.evaluate(test_data)
print(f"Random forest MAE: {results_forest[1]:.3f}")
RMSE = math.sqrt(results_forest[2])
print(f"Random forest RMSE: {RMSE:.3f}")
print("ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ")
print(f"Gradient boosting MAE: {results_gradient[1]:.3f}")
RMSE = math.sqrt(results_gradient[2])
print(f"Gradient boosting RMSE: {RMSE:.3f}")
```
